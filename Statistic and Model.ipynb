```python
rn = np.random.uniform(-2, 2, 30)
x=np.reshape(np.sort(rn,axis=0),(30,1))
```

```python
def Polynomial_function(Input):
    """This function will create a Polynomial quation 
       I have divided the function into 3 parts and then concate them.
    """
    angle1=np.pi*Input                   #  πx
    Radian1 =  (angle1*np. pi)/180       #  converting the angle into radian  (πx)*π/180
    #  -----------------------------------
    angle2=(3*np.pi*Input)/2             #  3πx/2
    Radian2= (angle2*np. pi)/180         #  converting the angle into radian
    # ---------------------------------------------
    expo=np.exp(-Input/2)                # e^-x/2
    function =expo*np.sin(Radian1) +np.sin(Radian2) #  f(x) = e^−x/2 sin(πx) + sin (3πx2)
    return function
y=Polynomial_function(x)
```

## Gradien_Descent:
<p>linear regression is a regression model that estimates the relationship between one independent variable and one dependent variable using a straight line. So to get the best regression line I have to find optimal weights and Baise and put them into the slope formula. in this task, I will also generate the Error of prediction with the help of the loss function.</p>

**Statistics:** I will start my Regression with 0 weight and baise and then calculate the rate of change of Weight and baise and then predict the outcomes. I am going to build the Gradient Descent with the help of these equations.



```python
def Gradient_descent(x,y):
    m_curr = 0  # Current Wight (slope) 
    b_curr = 0  # Current Baise (y_intercept)
    iterations = 200  # Number of iterations
    n=len(x)  # Number of points
    Weight_list =[] # This list will save the Weights of the model
    Baise_list = [] # This list will save the Baises of the model
    Loss_list = [] # This list will save the Loss of the model
    learning_rate = 0.001
    for i in range(iterations):
        y_predicted = m_curr*x + b_curr # Prediction = Weight *Points   + Baise
        cost = (1/n)*sum([val**2 for val in (y-y_predicted)]) # Loss = (1/N)*Σ (y-predicted)^2  
        md = -(2/n)*sum(x*(y-y_predicted)) # Change in Weights = (-2/N)*Σ* Points *(y-predicted)^2
        bd = -(2/n)*sum(y-y_predicted)     #  Change in Baise = (-2/N)*ΣN (y-predicted)^2
        m_curr = m_curr -learning_rate*md    #  Current weight - learning rate * Change in Weights
        Weight_list.append(m_curr)
        b_curr = b_curr-learning_rate*bd     #  Current Baise - learning rate * Change in Baise
        Baise_list.append(b_curr)
        Loss_list.append(cost)
        print(f"Weights {m_curr} , Baise {b_curr} , Loss {cost} , iterations {i}")
    return y_predicted , Weight_list,Baise_list,Loss_list
y_predicted,Weights,Baises,Loss=Gradient_descent(x,y)
```

<h1>Linear Regression Model</h1>

**Statistics:** first i have to calculate the Best weight and baise for model then i will predict the values


```python
def mean(values):
    """This function will find the mean of the data points"""
    return np.mean(values)
def variance(values, mean):
    """This takes to parameters values :data points andn mean of data points 
    and will calculate the total variace of points"""
    return sum([(x-mean)**2 for x in values]) # Variance = Σ(xi-mean of x) 
mean_X, mean_Y = mean(x), mean(y) 
var_x, var_y = variance(x, mean_X), variance(y, mean_Y)
```

```python
def covariance(x, mean_x, y, mean_y):
    """This function will take 4 parameters and will calculate the covariance of data points """
    covar = 0.0
    for i in range(len(x)):
        covar += (x[i] - mean_x)*(y[i] - mean_y)  #  Covariance = (x-mean of x)*(y- mean of y)
    return covar 
def coefficients(x,y):
    """This function will calculate the weight and baise for prediction"""
    x = x
    y = y
    mean_x, mean_y = mean(x), mean(y)
    weight = covariance(x, mean_x, y, mean_y) / variance(x, mean_x) # Weights =Covariance/variance
    baise = mean_y - weight * mean_x # Baise = mean of y - weight*mean of x
    return [weight, baise]
```

<h1>Let's predict the outcomes</h1>

```python
predictions = [] #  this list will contain the prediction points
weight, baise = coefficients(x,y) #  In this step we are calculating the wight and baise
for i in x:
    ynew =weight *i + baise  # y = mx + c
    predictions.append(ynew)
```

```python
regression_line = [(baise + weight*x) for x in x]
plt.scatter(x,predictions , c='b', label='Prediction')
plt.plot(x, regression_line, c='r', label='Regression Line')
plt.legend(loc='upper left')
plt.title('Linear Regression Model',fontweight="bold")
plt.show()
```

## Task:2 gaussian radial basis function (RBF):
<p>The radial distribution function (rdf) defines the probability of finding a particle at distance r from another tagged particle.So for Radial distribution I need a r distace from the center point.</p>

**Statistics:** First i will generate some random points for distribution and define the number of basis and sigmoid and then calculate the the r distance and then put them into the RBF equation which will distribute the data and i will calculate the weights by using these equation,





```python
rn = s = np.random.normal(5, 0.1, 30)
x=np.reshape(np.sort(rn,axis=0),(30,1))
y=Polynomial_function(x)
```

```python
num_basis = 5 # Defining the number of Basis 
sigma=0.1 

basis_idx = np.random.choice(x.shape[0],num_basis,replace =False) 
basis = np.array(x[basis_idx])
```

<h1>Plotting Gussain Basis</h1>

```python
Range = np.arange(np.min(x),np.max(x),0.01)
plt.figure(figsize=(8,5))
for mu in basis:
    plt.plot(Range,norm.pdf(Range,mu,sigma))
plt.title("Number of basis with 0.1 sigma",fontweight="bold")
plt.show()
```

<p>As you can see there are 5 basis.We can try different number of basis and sigma value for better results </p>

<h2>Building a radial basis function</h2>

```python
plt.figure(figsize=(8,5))
fig = plt.scatter(x,y,c='b',marker='o',label='Data Points') 
GBF = lambda x,mu : np.exp(-0.5 * (x - mu) ** 2 / sigma ** 2) # RBF = exp((-1/2)*((x-xj)/sigma)2)

phi = np.asmatrix(np.ones((x.shape[0],1)))  # creating a matrix of ones
for mu in basis:
    col = [GBF(obs,mu) for obs in x]
    phi = np.hstack((phi,col))  # Arranging combining the phi and GBF points and rearranging then horizontally

t = np.asmatrix(np.ones((Range.shape[0],1)))  # Creating the transpose matrix of x

for mu in basis:
    col = [GBF(obs,mu) for obs in Range]
    t = np.hstack((t,col)) # A transpose matrix of x
w = ((phi.T*phi)**-1)*phi.T*y # Weight = (xt*x)^-1 *sigma2
plt.plot(Range,t*w,c='y',marker='.',alpha=.8,label="GBF Distribution")
plt.title("GBF Distribution",fontweight="bold")
plt.legend(loc='upper left')
plt.show()
```

<p>As you can see GBF is doing great. i have set the sigma = 0.1  and choose 5 number of basis.So can try different sigma and basis but i think this the optimal value</p>

## Task:3 Classification by Fisher LDA

<p>In this code, i build a Fisher LDA model to create a boundary between classes and projected them on the same plane by maximize their mean difference and minimize their variences. In this code i perform following tasks: </p>
<p><b>1)  </b> I take the mean of their classes individualy.</p>
<p><b>2)  </b>I take the overall mean of class.</p>
<p><b>3)  </b>I calculate the mean sum  within classes by subtracting individual mean from the overall mean and take a dot  product with their Transpose. And sum the all values.</p>
<p><b>4)  </b>Then I calulate mean sum of overall classes by taking a difference of each class mean and overall classes and then take a dot product by their transpose and after that i multiply it with their number of classes and then i sum  up all the values.</p>
<p><b>5)  </b>In this step i take the inverse of sum within classes and take a dot product with sum of overall mean.</p>
<p><b>6)  </b>In this step i create the eigenvectors to project the classes on pi plane. and then i 
 transform the data for projection.</p>




```python
class FisherLDA():
    def __init__(self,n_components):
        self.n_components=n_components
        self.linear_discriminants=None
        self.eigenvalues = None
        self.eigenvectors = None
    def fit(self,X,y):
        n_features = X.shape[1]
        class_labels = np.unique(y)
        
        # s_w , s_b
        mean_overall = np.mean(X,axis=0)
        s_w = np.zeros((n_features,n_features))
        s_b = np.zeros((n_features,n_features))
        
        for c in class_labels:
            x_c = X[y==c]
            mean_c = np.mean(x_c,axis=0)
            s_w += (x_c-mean_c).T.dot(x_c-mean_c)
            
            n_c = x_c.shape[0]
            
            mean_diff = (mean_c - mean_overall).reshape(n_features,1)
            s_b += n_c*(mean_diff).dot(mean_diff.T)
        inv = np.linalg.inv(s_w).dot(s_b)
        self.eigenvalues , self.eigenvectors = np.linalg.eig(inv)
        self.eigenvectors = self.eigenvectors.T
        idx = np.argsort(abs(self.eigenvalues))[::-1]
        self.eigenvalues  =  self.eigenvalues[idx]   
        self.eigenvectors = self.eigenvectors[idx]
        
        self.linear_discriminants = self.eigenvectors[0:self.n_components]
    def transform(self,X):
        return np.dot(X,self.linear_discriminants.T)
```

```python
def getLoss(w,x,y,lam):
    m = x.shape[0] #First we get the number of training examples
    y_mat = oneHotIt(y) #Next we convert the integer class coding into a one-hot representation
    scores = np.dot(x,w) #Then we compute raw class scores given our input and current weights
    prob = softmax(scores) #Next we perform a softmax on these scores to get their probabilities
    loss = (-1 / m) * np.sum(y_mat * np.log(prob)) + (lam/2)*np.sum(w*w) #We then find the loss of the probabilities
    grad = (-1 / m) * np.dot(x.T,(y_mat - prob)) + lam*w #And compute the gradient for that loss
    return loss,grad


def oneHotIt(Y):
    m = Y.shape[0]
    #Y = Y[:,0]
    OHX = scipy.sparse.csr_matrix((np.ones(m), (Y, np.array(range(m)))))
    OHX = np.array(OHX.todense()).T
    return OHX

def softmax(z):
    z -= np.max(z)
    sm = (np.exp(z).T / np.sum(np.exp(z),axis=1)).T
    return sm

def getProbsAndPreds(someX):
    probs = softmax(np.dot(someX,w))
    preds = np.argmax(probs,axis=1)
    return probs,preds
```

```python
import scipy
w = np.zeros([X_train.shape[1],len(np.unique(y_train))])
lam = 1.5
iterations = 1000
learningRate = 0.001
losses = []
for i in range(0,iterations):
    loss,grad = getLoss(w,X_train,y_train,lam)
    losses.append(loss)
    w = w - (learningRate * grad)
print (loss)
```

```python
def getAccuracy(someX,someY):
    prob,prede = getProbsAndPreds(someX)
    accuracy = sum(prede == someY)/(float(len(someY)))
    return accuracy

print ('Training Accuracy: ', getAccuracy(X_train,y_train))
print ('Test Accuracy: ', getAccuracy(X_test,y_test))
```
